## **5.1.3 Data analysis and data visualisation**

```python
#load data
from matplotlib import pyplot as plt
import seaborn as sns
import folium
listings = pd.read_csv('data/airbnb_nyc.csv', delimiter=',')
```

```python
#let's check type for bedrooms

print(listings['bedrooms'].dtype)
```

```python
#let's get an idea of outlier situation - we can use a scatterplot to see this

sns.scatterplot(data=listings, x='bedrooms', y='price')
plt.title('Scatterplot showing showing price of Airbnb listings and number of bedrooms')
```

It would probably be worthwhile in this case to remove outliers - shown clearly in this scatterplot. Although using percentiles is possible, it may be easier in this case to filter/subset data.

```python
#subset and assign modified dataset to a new variable
listings[listings['bedrooms'] < 6]
listings_bedrooms_filtered = listings[listings['bedrooms'] < 6]
```

```python
#to get an idea of if/how/to what extent price changes depending on number of bedrooms we can do a kde plot

plt.figure(figsize =(15,10))
sns.kdeplot(data= listings_bedrooms_filtered, x='price', hue='bedrooms', palette='Dark2')
plt.title('Kde plot showing price of Airbnb listings and number of bedrooms')
```

This kde plot shows that as the number of bedrooms increase the price does somewhat increase. The fact that we see so many one bedroom flats is not surprising, given that this data is from such a big city. It is also worth noting that what we see here (in terms of number of bedrooms) is also only one factor contributing to price - and perhaps not as important as expected. There are of course also many other features that contribute to a higher price listing. Some of the one bedroom flats (perhaps the ones with high end furnishings/appliances/views/in best location etc.) are listed in a similar price-range to the listings with more bedrooms - this would explain what we see here.  

```python
#we can see what the actual correlation of these two features is

correlation = listings_bedrooms_filtered['price'].corr(listings_bedrooms_filtered['bedrooms'])
print(correlation)
```

This shows that bedroom number and price is positively correlated - moderately so. Possible reasons for this as mentioned above.

```python
#based on what we've observed, I would be curious to know what the descriptive stats (median in particular) show us
#a boxplot is a better way to visualise this than the plot above

plt.figure(figsize =(15,10))
sns.boxplot(data= listings_bedrooms_filtered, y='price', x='bedrooms', palette='CMRmap')
plt.title('Kde plot showing price of Airbnb listings and number of bedrooms')
```

This is interesting as it shows us clearly that the median (which is a measure of central tendency), in terms of price, does increase as bedroom number goes up. Although most of the one bedroom listings fall in the lower price range (as we saw in the kde plot), there are several outliers in terms of price - ones with, comparatively, high prices. This would explain why the correlation between bedroom number and price wasn't particularly strong.

```python
#The descriptive stats below reiterate what we have seen above - also, note the mean is higher than median - a result of outliers observed
listings_bedrooms_filtered[listings_bedrooms_filtered['bedrooms'] == 1]['price'].describe()
```
